```python
#Создайте вектор, скаляр, матрицу и тензор со значениями по вашему выбору, используя tf.constant().
vect = tf.constant([ 100, 200, 300])
scalarr = tf.constant(42)
Matrixx = tf.constant([[4, 5], [40, 50]])
tensorr = tf.constant([
    [[4, 5], [40, 50]],
    [[3, 1], [43, 52]],
    [[6, 2], [10, 40]]
    ])
vect, scalarr, Matrixx, tensorr
```

```python
#Найдите форму, ранг и размер тензоров, которые вы создали в 1.
@tf.function
def tf_size_func(tensorr):
  print("Shape ", tensorr.shape)
  print("Rank ", tensorr.ndim)
  print("Size ", tf.size(tensorr))
  print("|||||||||||||||||||||||||||||||||")

tf_size_func(vect)
tf_size_func(scalarr)
tf_size_func(Matrixx)
tf_size_func(tensorr)
```

```python
#Создайте два тензора, содержащих случайные значения от 0 до 1, с помощью shape [5, 300].
tensor_1 = tf.random.uniform(shape=[5, 300])
tensor_2 = tf.random.uniform(shape=[5, 300])
# Print the tensors
print("Tensor 1:")
print(tensor_1)
print("Tensor 2:")
print(tensor_2)
```

```python
#Умножьте два тензора, которые вы создали в 3, используя матричное умножение.
tf.matmul(tf.transpose(tensor_1), tensor_2)
```

```python
#Умножьте два тензора, которые вы создали в 3, используя скалярное произведение.
tf.tensordot(tf.transpose(tensor_1), tensor_2, axes=1)
```

```python
#Создайте тензор со случайными значениями от 0 до 1 с помощью shape [224, 224, 3].
tensor_3 = tf.random.uniform(shape=[224, 224, 3])
tensor_3
```

```python
#Найдите минимальное и максимальное значения тензора, созданного вами в разделе 6.
print("min = ", tf.reduce_min(tensor_3).numpy())
print("max = ", tf.reduce_max(tensor_3).numpy())
```

```python
#Создайте тензор со случайными значениями формы [1, 224, 224, 3], затем сожмите его, чтобы изменить форму на [224, 224, 3].
tensor_4 = tf.random.uniform(shape=[1, 224, 224, 3])
tensor_4
```

```python
#9.Создайте тензор с помощью shape [10], используя выбранные вами значения, затем найдите индекс, который имеет максимальное значение.
C = random_tensor = tf.random.uniform(shape=[10], minval=0, maxval=10, dtype=tf.int32)
print(tf.argmax(C).numpy())
C
```

```python
#10.One-hot encode the tensor you created in 9.
num_classes = len(tf.unique(C)[0])

# One-hot кодирование
C_one_hot = tf.one_hot(indices=C, depth=num_classes)

C_one_hot
```
